```python
import sys
from pathlib import Path

project_root = Path("MS1-Based-Workflow").resolve()
src_path = project_root / "src"

if str(src_path) not in sys.path:
    sys.path.insert(0, str(src_path))

print("Connected to:", project_root)
```

```python
mz_dial_file = project_root / "examples" / "mzdial_example.xlsx"
standards_file = project_root / "examples" / "standards_example.xlsx"

print("mzDial file exists:", mz_dial_file.exists())
print("Standards file exists:", standards_file.exists())
```

```python
mz_dial_data = load_file(mz_dial_file)
standards_data = load_file(standards_file)

print("mzDial shape:", mz_dial_data.shape)
print("Standards shape:", standards_data.shape)

mz_dial_data.head()
```

Run without standards

```python
workflow = MS1Workflow(
    ppm_threshold=5,
    rt_order_tolerance=0.005,
    output_dir=project_root / "outputs"
)

results_without = workflow.run_without_standards(
    mz_dial_data=mz_dial_data
)
```

Run with standards

```python
workflow = MS1Workflow(
    ppm_threshold=5,
    rt_order_tolerance=0.005,
    rt_alignment_tolerance=0.01,
    output_dir=project_root / "outputs"
)

results_with = workflow.run_with_standards(
    mz_dial_data=mz_dial_data,
    standards=standards_data
)
```

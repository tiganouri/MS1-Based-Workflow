# MS1-Based Workflow Demo

This notebook demonstrates how to run the MS1-based annotation workflow using the example datasets provided in this repository.

The workflow supports two modes:

1. **Without standards**
2. **With standards**

```python
import sys
from pathlib import Path

project_root = Path("MS1-Based-Workflow").resolve()
src_path = project_root / "src"

if str(src_path) not in sys.path:
    sys.path.insert(0, str(src_path))

print("Connected to:", project_root)
```

```python
mz_dial_file = project_root / "examples" / "mzDial_MS1_example.xlsx"
standards_file = project_root / "examples" / "standards_example.xlsx"

print("mzDial file exists:", mz_dial_file.exists())
print("Standards file exists:", standards_file.exists())
```

Run without standards

```python
from ms1_workflow.io import load_file
from ms1_workflow.workflow import MS1Workflow

NIAS_file = project_root / "examples" / "NIAS_example.xlsx"
NIAS_data = load_file(NIAS_file)

workflow = MS1Workflow(
    ppm_threshold=5,
    rt_order_tolerance=0.005,
    output_dir=project_root / "outputs"
)

results_without = workflow.run_without_standards(
    mz_dial_data=NIAS_data
)
```

Run with standards

```python
PET_file = project_root / "examples" / "PET_example.xlsx"
PET_data = load_file(PET_file)

standards_data = load_file(standards_file)

workflow = MS1Workflow(
    ppm_threshold=5,
    rt_order_tolerance=0.005,
    rt_alignment_tolerance=0.01,
    output_dir=project_root / "outputs"
)

results_with = workflow.run_with_standards(
    mz_dial_data=PET_data,
    standards=standards_data
)
```
